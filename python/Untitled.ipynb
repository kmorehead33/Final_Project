```python
import pandas as pd
import numpy as np
```

```python
results_db = pd.read_csv("../data/game_data.csv")
results_db.tail(10)
```

```python
def underdog_score(game):
    if game['underdog'] == game['team_home']:
        return game['score_home']
    else:
        return game['score_away']

results_db['score_underdog'] = results_db.apply(underdog_score, axis=1)
    
```

```python
def favorite_score(game):
    if game['favorite'] == game['team_home']:
        return game['score_home']
    else:
        return game['score_away']

results_db['score_favorite'] = results_db.apply(favorite_score, axis=1)
    
```
