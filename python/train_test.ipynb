```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt 
```

```python
results_db = pd.read_csv("../data/train_test_data.csv")
```

```python
results_db = results_db[['favorite', 'underdog', 'spread_favorite', 'season_total_points_for _favorite', 'season_total_points_against_favorite', 'season_total_points_for _underdog', 'season_total_points_against_underdog', 'home_win', 'favorite_win']]
```

```python
results_db = results_db.rename(columns={"season_total_points_for _favorite": "season_total_points_favorite", "season_total_points_for _underdog": "season_total_points_underdog"})
```

```python
results_db.head()
```

```python
team_mapping = {}

for idx, row in results_db.iterrows():
    fave_name = row["favorite"]
    if fave_name not in team_mapping:
        team_mapping[fave_name] = len(team_mapping)
    under_name = row["underdog"]
    if under_name not in team_mapping:
        team_mapping[under_name] = len(team_mapping)

print(team_mapping)
```

```python
for name in team_mapping:
    results_db.replace(to_replace=name, value=team_mapping[name], inplace=True)
results_db.head()
```

```python
#split data into training and testing 
```

```python
y=results_db["favorite_win"].values.reshape(-1,1)
X=results_db.drop(columns="favorite_win")
X.shape

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

```python
#create a logical regression model
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier

classifier.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
#make predictions
```

```python
predictions = classifier.predict(X_test)
y_list = [x[0] for x in y_test]
print(f"First 10 Predictions:   {predictions[:10]}")
print(f"First 10 Actual labels: {y_list[:10]}")
```

```python
pd.DataFrame({"Prediction": predictions, "Actual": y_list}).reset_index(drop=True)
```
