```python
import pandas as pd
import numpy as np
```

```python
results_db = pd.read_csv("../data/game_data.csv")
results_db.tail(10)
```

```python
def underdog_score(game):
    if game['underdog'] == game['team_home']:
        return game['score_home']
    else:
        return game['score_away']

results_db['score_underdog'] = results_db.apply(underdog_score, axis=1)
    
```

```python
def favorite_score(game):
    if game['favorite'] == game['team_home']:
        return game['score_home']
    else:
        return game['score_away']

results_db['score_favorite'] = results_db.apply(favorite_score, axis=1)
    
```

```python
results_db["favorite_win"] = [1 if x>y else 0 for x,y in zip(results_db['score_favorite'], results_db['score_underdog'])]
```

```python
season_2015 = results_db['schedule_season'] == 2015
season_2016 = results_db['schedule_season'] == 2016
season_2017 = results_db['schedule_season'] == 2017
season_2018 = results_db['schedule_season'] == 2018
```

```python
season_2015_db = results_db[season_2015]
season_2016_db = results_db[season_2016]
season_2017_db = results_db[season_2017]
season_2018_db = results_db[season_2018]
```
