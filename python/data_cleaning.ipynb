```python
import pandas as pd
import numpy as np
```

```python
results_db = pd.read_csv("../data/results.csv")
results_db.tail(1000)
```

```python
results_db = results_db[["schedule_date", "schedule_season", "schedule_week", "team_home", "score_home", "score_away", "team_away", "team_favorite_id", "spread_favorite"]]
```

```python
results_db.head()
```

```python
after_2015 = results_db["schedule_season"]>= 2015
```

```python
results_db = results_db[after_2015]
```

```python
results_db.head()
```

```python
before_2019 = results_db["schedule_season"]< 2019
```

```python
results_db = results_db[before_2019]
```

```python
results_db.tail(10)
```

```python
results_db["home_win"] = [1 if x>y else 0 for x,y in zip(results_db['score_home'], results_db['score_away'])]
```

```python
results_db["away_win"] = [1 if x<y else 0 for x,y in zip(results_db['score_home'], results_db['score_away'])]
```

```python
results_db.head()
```

```python
results_db.to_csv("../data/game_data.csv", index=False)
```
